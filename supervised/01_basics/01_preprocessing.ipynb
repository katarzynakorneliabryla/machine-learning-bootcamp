### <a name='0'> </a> Import bibliotek

```python
import numpy as np
import pandas as pd
import sklearn
```

### <a name='1'> </a> Wygenerowanie danych

```python
data = {
    'size' : ['XL', 'L', 'M', 'L', 'M'],
    'color' : ['red', 'green', 'blue', 'green', 'red'],
    'gender' : ['female', 'male', 'male', 'female', 'female'],
    'price' : [199.0, 89.0, 99.0, 129.0, 79.0],
    'weight' : [500, 450, 300, 380, 410],
    'bought' : ['yes', 'no', 'yes', 'no', 'yes']
}

df_raw = pd.DataFrame(data=data)
df_raw
```

### <a name='2'> </a> Utworzenie kopii danych

```python
df = df_raw.copy()
df.info()
```

### <a name='2'> </a> Zmiana typu danych i wstÄ™pna eksploracja

```python
for col in['size', 'color', 'gender', 'bought']:
  df[col] = df[col].astype('category') # zmiana na dane kategoryczne

df['weight'] = df['weight'].astype('float')

df.info()
```

```python
df.describe().T # domyÅ›lnie - tylko dla kolumn numerycznych
```

```python
df.describe(include=['category']).T
```

### <a name='3'> </a> LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
le.fit(df['bought'])
le.transform(df['bought'])
```

```python
le.fit_transform(df['bought']) # zamiast osobno fit i transform
```

```python
le.classes_ # mÃ³wi nam jak wyglÄ…da mapowanie
```

```python
df['bought'] = le.fit_transform(df['bought'])
df
```

```python
le.inverse_transform(df['bought']) # wraca do oryginalnej postaci
```

```python
df['bought'] = le.fit_transform(df['bought'])
df
```

### <a name='4'> </a> OneHotEncoder

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder(sparse=False)
encoder.fit(df[['size']])
```

```python
encoder.transform(df[['size']])
```

```python
encoder.categories_
```

```python
encoder = OneHotEncoder(drop='first', sparse=False)
encoder.fit(df[['size']])
encoder.transform(df[['size']])
```

### <a name='5'> </a> Pandas *get_dummies()*

```python
pd.get_dummies(data=df) # zmienne kategoryczne sÄ… kodowane
```

```python
pd.get_dummies(data=df, drop_first=True) # zmienne kategoryczne sÄ… kodowane z usuwaniem pierwszej kolumny
```

```python
pd.get_dummies(data=df, drop_first=True, prefix='new')
```

```python
pd.get_dummies(data=df, drop_first=True, prefix_sep='-')
```

```python
pd.get_dummies(data=df, drop_first=True, columns=['size']) # kodowanie tylko jednej kolumny
```

### <a name='6'> </a> Standaryzacja - StandardScaler

std() - pandas nieobciÄ…Å¼ony

std() - numpy obciÄ…Å¼ony

```python
print(f"{df['price']}/n")
print(f"Åšrednia: {df['price'].mean()}")
print(f"Odchylenie standardowe: {df['price'].std():.4f}")
```

```python
(df['price'] - df['price'].mean()) / df['price'].std()
```

```python
def standarize(x):
  return (x - x.mean()) / x.std()

standarize(df['price'])
```

```python
from sklearn.preprocessing import scale

scale(df['price'])
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(df[['price']])
scaler.transform(df[['price']])
```

```python
scaler = StandardScaler()
df[['price', 'weight']] = scaler.fit_transform(df[['price', 'weight']])
```

### <a name='7'> </a> Przygotowanie danych do modelu

```python
df = df_raw.copy()
df
```

```python
le = LabelEncoder()

df['bought'] = le.fit_transform(df['bought'])

scaler = StandardScaler()
df[['price', 'weight']] = scaler.fit_transform(df[['price', 'weight']])

df = pd.get_dummies(data=df, drop_first=True)
df
```
